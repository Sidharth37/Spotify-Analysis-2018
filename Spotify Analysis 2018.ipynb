```python
#importing data set
import pandas as pd
filename="C:/Users/Sidharth Jahagirdar/Desktop/First Sem/Spotify_2018/spotifyclassification/data.csv"
df=pd.read_csv(filename,encoding='ISO_8859_1')
df.head(10)
```

```python
#import libraries needed to perform regression analysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
#Data Cleaning
new_order=[2,1,4,5,7,8,11,10,13]
df=df[df.columns[new_order]]
```

```python
#Defining dependent and independent variables
x=df.iloc[:,1:].values
y=df.iloc[:,0:1].values
```

```python
#Splitting data into train and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.05)
```

```python
#Creating Regression Model
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Creating trial data using Test data example of test data index 6 is shown here
trial_data=x_test[29]
trial_data=trial_data.reshape(-1,8)
trial_data=trial_data.astype(float)
```

```python
#Predicting values on test data
regressor.predict(trial_data)
```
